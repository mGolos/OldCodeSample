```python
FC0 = Tf.fPearsonCorrelation(patts0)
FC1 = Tf.fPearsonCorrelation(patts1)
```

```python
Td.mapMatrices([FC0,FC1], fig=figure(figsize=((15,7))), lign=1)
```

```python
Td.mapMatrices([SS0,SS1], fig=figure(figsize=((15,4))), aspect='auto', interpolation='nearest', lign=1)
```
